````python
''' As per professor - for yasmine we need to generate serveral networks
   each one a RD network.

    num_nodes and num_networks define the size of the RD networks
    in the paper its 300 networks in total, the networks each with 45
    nodes. -- this is to see if we are on the right path, and graphs
    that size just look like a mess, so I lowered the size so we can see
    the networks, so we can see that we are on the right path.

    [ Note there are also the task attributes, we havent addressed that yet.]

'''

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# global ps

def generate_er_networks(num_nodes, num_networks, names):
    global ps
    ps = np.linspace(0.2, 1.0, num_networks)
    networks = []
    for i, p in enumerate(ps):
        G = nx.erdos_renyi_graph(num_nodes, p)
        node_mapping = {i: name for i, name in enumerate(names)}
        G = nx.relabel_nodes(G, node_mapping)
        networks.append(G)
    return networks

'''
def read_names_from_file(file_path):
    with open(file_path, 'r') as file:
        names = [name.rstrip('\n') for name in file.readlines()]
    return names
'''
def plot_er_networks(networks):
    num_networks = len(networks)
    fig, axs = plt.subplots(3, 3, figsize=(20, 20))
    for i, ax in enumerate(axs.flat):
        if i < num_networks:
            nx.draw(networks[i], ax=ax, with_labels=True, node_color='lightblue', node_size=1000, font_size=8)
        ax.set_title(f"p = {ps[i]:.2f}")
        ax.set_xticks([])
        ax.set_yticks([])
    plt.tight_layout()
    plt.show()

# Example usage
# Yasmin changed the nodes from 6 to 30
# Yasmin changed the networks from 9 to 25
num_nodes = 15  # Number of nodes in each network
num_networks = 20  # Number of networks to generate
names_file_path = 'names.txt'  # File path containing the names, one name per line

#names = read_names_from_file(names_file_path)
names = names_list
er_networks = generate_er_networks(num_nodes, num_networks, names)
plot_er_networks(er_networks)
````

````python
####### Dr. Esterline's Code #######################################
# -*- coding: utf-8 -*-
"""
Created on Fri Dec 23 21:01:59 2022

@author: Dr. Albert Esterline
"""

####### TODO
####        get the speed, accurac, reliablity data from
####        the network decorate nodes
import networkx as nx
from numpy import random

task_list = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']

def decorate_nodes(G):
    """Add randomly generated values for all 4 attribute for all nodes of G.

    For each node n, randomly generate values for attributes accuracy, speed,
    and reliability, for each, from a uniform [0.0, 1.0) distribution. Also,
    for attribute tasks, randomly select a random number in [2,5] of tasks
    from the global task list.

    Parameters
    ----------
    G : NetworkX Graph
        Each node of G has random values assigned to its 4 attributes accuracy,
        speed, reliability, and tasks

    Returns
    ------
    None
    """
    for n in G.nodes():
        # Get 3 random numbers from the uniform distribution [0, 1).
        # Normalize them so their sum is 1.0. Use them for, respectively,
        # the values of the attribues accuracy, speed,  reliability of node n

        accuracy, speed, reliability = random.uniform(size=3)

        G.nodes()[n]['accuracy'] = accuracy
        G.nodes()[n]['speed'] = speed
        G.nodes()[n]['reliability'] = reliability

        # Get a random number between 2 and 5 for the size of the list of tasks
        # assigned to the tasks attribute of n. There are x tasks from
        # task_list randomly chosen without replacement.
        x = random.randint(2,6)
        G.nodes()[n]['tasks'] = \
            list(random.choice(task_list, size=x, replace=False))

#nx.write_gpickle(G, 'G_pickled')

# Generate a network based on P -- where p is ??
def gen_network(p):
    G = nx.erdos_renyi_graph(30, p) # 30 nodes and p
    decorate_nodes(G)
    return G

#################################################################

def total_rank(d):
    """ Given a dictionary d, return a list of keys of d ordered in
    non-increasing order of their associated values.
    """
    from operator import itemgetter

    ss = sorted(d.items(), key=itemgetter(1), reverse=True)
    return [x for x, _ in ss]

#################################################################

import functools

def tasks_covered(EG, tasks):
    """ Return True iff all the tasks in argument tasks are covered by the task
    lists that are the values of the tasks attributes of the nodes of G."""
    # Get a list lis of the tasks lists of all nodes of G. Return True iff the
    # tasks in argument tasks is a subset of the union of the task lists in lis.
    lis = [EG.nodes()[n]['tasks'] for n in EG.nodes()]
    return set(tasks) <= set(functools.reduce(lambda a, b: set(a) | set(b), lis))

##########################################################

def mean_centrality(G):
    bc = nx.betweenness_centrality(G)

    # Get the normalized eigenvector centrality
    ec = nx.eigenvector_centrality(G)
    sum_ec = sum(ec.values())
    ec_normalized = {k : v/sum_ec for k,v in ec.items()}

    # Return the harmonic mean of the betweenness centrality and eigenvector
    # centrality for each node
    return {k: (bc[k]*ec_normalized[k])**0.5 for k in ec_normalized.keys()}


def relative_rank(n, sorted_nodes):
    """ Return the rank of n in sorted_nodes. The highest rank is 1.0, the
    lowest, 0.0.
    """
    return 1.0 - sorted_nodes.index(n) / (len(sorted_nodes) - 1)

def average_rank(EG, sorted_nodes):
    return sum([relative_rank(n, sorted_nodes) for n in EG.nodes()]) / len(EG)

def combed_centrality_rank(EG, G, n):
    sorted_nodes = total_rank(mean_centrality(G))
    return (relative_rank(n,sorted_nodes) + average_rank(EG,sorted_nodes)) / 2

#########################################################

# The following function returns the core number of node n in Graph G
# normalized by dividing by the largest core number of any node in G.
def relative_core_num(G, n, max_core_num):
    """ Return the core number of node n in Graph G normalized by dividing by
    the largest core number of any node in G."""
    CN = nx.core_number(G)
    return CN[n] / max_core_num

# The following functions return, respectively, the average core number of the
# nodes in Graph EG and the minimum core number of any node in EG, both
# normalized by dividing by the largest core number of any node in G.
def average_k(EG, G, max_core_num):
    """ Return the average core number of the nodes in Graph EG normalized by
    dividing by the largest core number of any node in G. """
    CN = nx.core_number(EG)
    return (sum(CN.values()) / len(CN)) / max_core_num

def min_k(EG, G, max_core_num):
    """ Return the minimum core number of the nodes in Graph EG normalized by
    dividing by the largest core number of any node in G. """
    return min(nx.core_number(EG).values()) / max_core_num

def combed_core_num(EG, G, n):
    max_core_num = max(nx.core_number(G).values())
    if max_core_num == 0: max_core_num = 1
    rel = relative_core_num(G, 0, max_core_num)
    ave = average_k(EG, G, max_core_num)
    mn  = min_k(EG, G, max_core_num)
    return (rel + ave + mn) / 3.0

#########################################################
import math

def min_max_all_attrs(G, attrs):
    def min_max_attr(G, attr):
        mn, mx = 1, 0
        present = False
        for n in G.nodes():
            if attr in G.nodes()[n]:
                present = True
                val = G.nodes()[n][attr]
                if val > mx: mx = val
                if val < mn: mn = val
        if not present:
            raise Exception('Attribute ' + attr +
                            ' not present in the network')
        return mn, mx

    return {attr: min_max_attr(G, attr) for attr in attrs}

def ave_all_attr_values_norm(EG, attrs, d):
    def ave_attr_value_norm(EG, attr, mx):
        sm, cnt = 0.0, 0
        for n in EG.nodes():
            if attr in EG.nodes()[n]:
                sm += EG.nodes()[n][attr]
                cnt += 1
        return (sm/cnt)/mx if cnt > 0 else 0

    return sum([ave_attr_value_norm(EG, attr, d[attr][1]) for attr in attrs])\
           / len(attrs)

def g_mean_min_attr_values_norm(EG, attrs, d):
    def min_attr_value_norm(EG, attr, mn, mx):
        mn1 = 1.0
        for n in EG.nodes():
            if attr in EG.nodes()[n]:
                val = EG.nodes()[n][attr]
                if val < mn1: mn1 = val
        # Guard against multiplication by 0 with a very small value instead of 0
        #diff = 0.0005 if mn1 == mn else mn1 - mn
        #print((0.003 + 0.997 * (diff / (mx - mn))) ** (1/5))
        return 1.0 if mx == mn else (0.003 + 0.997 * ((mn1 - mn) / (mx - mn))) ** (1/5)

    return math.prod([min_attr_value_norm(EG, attr, *d[attr])
                      for attr in attrs]) ** (1/len(attrs))
    # return len(attrs) / \
    #     sum([1 / min_attr_value_norm(EG, attr, *d[attr]) for attr in attrs])

def ave_min_all_attr_values_norm(EG, G, attrs):
    global_dict = min_max_all_attrs(G, attrs)
    return ave_all_attr_values_norm(EG, attrs, global_dict), \
           g_mean_min_attr_values_norm(EG, attrs, global_dict)


#########################################################

def average_assortativity(EG, attrs):
    """ Return the average of the assortativity for EG over attributes accuracy,
    speed, and reliability"""
    # Get a list of the assortativity of G for attributes accuracy, speed, and
    # reliability, and return their average.
    assort = [nx.numeric_assortativity_coefficient(EG,x) for x in attrs]
    return ((sum(assort) / 3.0) + 1) / 2

##############################################################

# The following functions return, respectively, the average degree of the
# nodes in graph G and the minimum degree of any node in G.
def average_degree(EG, G, max_deg):
    """ Return the normalized average degree of the nodes in Graph EG. """
    return (sum(dict(G.degree(nbunch=EG.nodes())).values()) / len(EG)) / max_deg

def min_degree(EG, max_deg):
    """ Return the normalized minimum cdegree of the nodes in Graph EG. """
    return min(dict(EG.degree()).values()) / max_deg

def combed_degree(EG, G):
    max_deg = max(dict(G.degree()).values())
    if max_deg == 0: max_deg = 1
    return (average_degree(EG, G, max_deg) + min_degree(EG, max_deg)) / 2.0

###############################################################

def normed_combed_deg_assort(G, EG):
    deg_assort_full = nx.degree_assortativity_coefficient(G)
    deg_assort_embed_ego = \
        nx.degree_assortativity_coefficient(G, nodes=EG.nodes())
    deg_assort_ego = nx.degree_assortativity_coefficient(EG)
    deg_assort_embed_ego_norm = (deg_assort_embed_ego - deg_assort_full) /2
    deg_assort_ego_norm = (deg_assort_ego - deg_assort_full) /2
    return (((deg_assort_embed_ego_norm + deg_assort_ego_norm) / 2.0) + 1) / 2

###################################################################

def overall(EG, G, n):
    attrs = ['accuracy', 'speed', 'reliability']
    ave_perf, min_perf = ave_min_all_attr_values_norm(EG, G, attrs)
    return 0.35 * combed_centrality_rank(EG, G, n) \
         + 0.23 * combed_core_num(EG, G, n)        \
         + 0.15 * ave_perf \
         + 0.09 * min_perf \
         + 0.07 * average_assortativity(EG, attrs)        \
         + 0.06 * combed_degree(EG, G)             \
         + 0.04 * normed_combed_deg_assort(G, EG)

##################################################################
import statistics
import numpy as np

#bkd --- store the networks we make here, in this list
glist = []

def global_ranking(G, tasks):
    fails = 0
    ranking = {}
    for n in G.nodes():
        EG = nx.ego_graph(G, n)
        if tasks_covered(EG, tasks):
            ranking[n] = round(overall(EG, G, n), 3)
        else:
            print("Node", n, " and its neighbors don't handle all tasks.")
            fails += 1
    return ranking, fails


# Run Trials - this may generate the networks
# - start
# - stp
# - numb ??
# - repititions?

def run_trials(strt, stp, num, reps):
    ps, ave_mins, ave_maxs, ave_aves, ave_fails = [], [], [], [], []
    for p in np.linspace(strt, stp, num=num):
        mins, maxs, aves, fails = [], [], [], []
        # for count in Repititions - for each reptition
        for cnt in range(reps):
            G = gen_network(p)
            # bkd - store this in our network list
            glist.append(G)
            rnking, fls = global_ranking(G, {'B', 'D', 'F'})
            fails.append(fls)
            vals = sorted(rnking.values())
            mins.append(vals[0])
            maxs.append(vals[-1])
            aves.append(statistics.mean(vals))
        ps.append(p)
        ave_fails.append(statistics.mean(fails))
        ave_mins.append(statistics.mean(mins))
        ave_maxs.append(statistics.mean(maxs))
        ave_aves.append(statistics.mean(aves))
    return ps, ave_mins, ave_maxs, ave_aves, ave_fails

pp, a_mins, a_maxs, a_aves, a_fails = run_trials(0.2, 1.0, 25, 12)

import matplotlib.pyplot as plt
# plt.plot(pp, a_mins, 'k^',  label='Minimum')
# plt.plot(pp, a_aves, 'k-',  label='Average')
# plt.plot(pp, a_maxs, 'k--', label='Maximum')
# plt.legend(fontsize=8)
# plt.xlabel('p (probability of an edge between any 2 nodes)')
# plt.ylabel('score', fontsize=15)
# plt.title('Score vs. p', fontsize=16)
# plt.figure()
plt.plot(pp, a_fails)
plt.xlabel('p (probability of an edge between any 2 nodes)')
plt.ylabel('Average failures for 45 nodes')
plt.title('Failures to cover all tasks vs. p')

# for i = 1 to 25
# for x in range(25):
#   run a round pp[x], 3
#    print(round(pp[x],3), round(a_mins[x],3), round(a_aves[x],3), round(a_maxs[x],3))

# delta_p =  0.7 - 0.332
# delta_min = 0.451  - 0.385
# delta_ave = 0.563 - 0.511
# delta_max = 0.676 - 0.629
# print(round(delta_min/delta_p, 3), round(delta_ave/delta_p, 3),
#       round(delta_max/delta_p, 3))

# def intersct(m, p0, y0):
#     return round(y0 - m * p0, 3)
# intersct(0.179, 0.332, 0.385)
# intersct(0.141, 0.332, 0.511)
# intersct(0.128, 0.332, 0.629)

# f = open('temp_vals_pp_v3.txt', 'w')
# f.write(str(pp))
# f.close()

# f = open('temp_vals_mins_v3.txt', 'w')
# f.write(str(a_mins))
# f.close()

# f = open('temp_vals_aves_v3.txt', 'w')
# f.write(str(a_aves))
# f.close()

# f = open('temp_vals_maxs_v3.txt', 'w')
# f.write(str(a_maxs))
# f.close()

# pp1 = pp[3:24]
# a_mins1 = a_mins[3:24]
# a_aves1 = a_aves[3:24]
# a_maxs1 = a_maxs[3:24]
# import pandas as pd
# from statsmodels.formula.api import ols
# data = pd.DataFrame({'x': pp1, 'y': a_mins1})
# model = ols("y ~ x", data).fit()
# print(model.params)
# data1 = pd.DataFrame({'x': pp1, 'y': a_aves1})
# model1 = ols("y ~ x", data1).fit()
# print(model1.params)
# data2 = pd.DataFrame({'x': pp1, 'y': a_maxs1})
# model2 = ols("y ~ x", data2).fit()
# print(model2.params)
````

## Class WebIDData - we can use this to store the intermediate data while we convert from NX graph to WebID


````python
#------------ WebID Data Class--------------------------------------------------------
#
class WebidData:
  def __init__(self, name='', accuracy=0, speed=0, reliability=0, tasks=[], friends=[]):
    self.name = name
    self.speed = speed
    self.reliability = reliability
    self.tasks = tasks
    self.friends = friends

  def __str__(self):
    class_name = type(self).__name__
    attributes = ", ".join(f"{attr}={value}" for attr, value in self.__dict__.items())
    return f"{class_name}({attributes})"

````

````python
#------------nxgraphs_to_dictionaries ------------------------------------------
## take a list of networks
## and convert them to a list of dictionaries
##  so we can turn that into RDF files
#
def nxgraphs_to_dictionaries(er_networks):
  all_networks = []
  i = 0
  for n in er_networks:
      friends = {}
      i = i + 1
      ## DEBUG print
      #print ('--------------------------------------------- ')
      #print (n)
      d = nx.to_dict_of_dicts(n).items()
      for key,val in d:
          key = str(key)  # if key is not a string, make it so
          #debug print
          #print (key," --- ", end='')
          # create a new dictionary to hold a person as a key and the value as their list of friends
          friends[key] = []
          for k in val:
              k = str(k)  # if the values are not strings, make them strings
              #debug print
              #print (k, ", ", end = '')
              # since friends value is a list of friends, append this new person to that list
              friends[key].append(k)
          #print("")
      all_networks.append(friends)
  print ("\n\n** Last network as a dictionary of friends **")
  print (friends)
  print ('----------------------------------------------------------------------------------')
  print ('    This is a dictionary representation of all of the ER networks produced        ')
  print ('      person :   [ their friend list ]')
  print ('      this we can use to populate the rdfs to create the servers')
  print ('----------------------------------------------------------------------------------')

  # return the list
  return all_networks

  ## debug
  ## import pprint
  ## pp = pprint.PrettyPrinter(indent=4)
  ## pp.pprint (all_networks)
````

````python
#------------nxgraphs_to_dictionaries2 ------------------------------------------
#------------------VERSION 2-----------------------------------------------------
#--  this version ALso gets the attributes (speed, accuracy,etc) from the network
#--  it also uses a class called WebidData - that has members for all the attributes
#--  this will make the programming of the writing to files easier to read.

## take a list of networks
## and convert them to a list of dictionaries
##  so we can turn that into RDF files
#
def nxgraphs_to_dictionaries2(er_networks):
  all_networks = []
  i = 0
  for n in er_networks:
      this_network = []
      i = i + 1
      # turn the network into a dictionary we can parse
      d = nx.to_dict_of_dicts(n).items()

      # for each person in the network, first get the count, then their
      #  name (in key), and their friends (in val)
      for count,(key,val) in enumerate(d):
          # lets make a person and give them attributes and friends
          aPerson = WebidData(str(key)) # add a person and make their name 'key'
          # unncessary initialization
          aPerson.reliability = 0
          aPerson.speed = 0
          aPerson.accuracy = 0
          aPerson.tasks = []
          aPerson.friends = []
          # end unnecessary init
          # n.nodes()[1] - this is how to get the attributes stored in the
          #  network data structure. Count - matches where we are in the list
          #  we pass they key, and it gives us the value, which we store in our person
          aPerson.accuracy = n.nodes()[count]['accuracy']
          aPerson.speed = n.nodes()[count]['speed']
          aPerson.reliability = n.nodes()[count]['reliability']
          aPerson.tasks = n.nodes()[count]['tasks']
          # now add friends, for each friend in that list, append to their friend list
          for ffriend in val:
             aPerson.friends.append(str(ffriend))

          # now finally append this network to our list of networks
          this_network.append (aPerson)
          #print ('DEBUG:', vars(aPerson))
      all_networks.append(this_network)
  print ("\n\n** Last network as a dictionary of friends **")
  print (friends)
  print ('----------------------------------------------------------------------------------')
  print ('    This is a dictionary representation of all of the ER networks produced        ')
  print ('      person :   [ their friend list ]')
  print ('      this we can use to populate the rdfs to create the servers')
  print ('----------------------------------------------------------------------------------')

  # return the list
  return all_networks

  ## debug
  ## import pprint
  ## pp = pprint.PrettyPrinter(indent=4)
  ## pp.pprint (all_networks)
````

````python
# ------ read_in_a_network ------------------------GET ATTRIBUTES RIGHT----------------------
# Turn a folder of RDF's into a network (G)
#
# input: a folder name, containing a set of RDF's to make a network
# output: a networkX graph

def read_in_a_network(folder_path):
  import rdflib
  import networkx as nx
  import matplotlib.pyplot as plt
  from urllib.parse import urlparse

  # Create a NetworkX graph object
  new_graph = nx.Graph()

  # List of TTL files
  # Get a list of files in the folder
  file_list = os.listdir(folder_path)
  network = []
  # lets go through each file and build our network
  for file in file_list:
      g = rdflib.Graph()

      file_name = folder_path + '/' + file
      g.parse(file_name, format='turtle')
      subjects = sorted(set(str(subj) for subj, _, _ in g))

      new_p = WebidData()
      new_p.friends = []
      new_p.tasks = []

      # We iterate over the graph and fill in the data
      count = 0
      for subj in subjects:
          for _, pred, obj in g.triples((rdflib.URIRef(subj), None, None)):

              pred_str = str(pred)
              obj_str = str(obj)
               # the first time through get the name from that first line
              if count == 0:
                parsed_url = urlparse(subj)
                new_p.name = parsed_url.netloc.split('.')[0]
              # get other attributes
              elif pred_str == 'http://ncat.edu/custom/accuracy':
                new_p.accuracy = float(obj_str)
              elif pred_str == 'http://ncat.edu/custom/reliability':
                new_p.reliability = float(obj_str)
              elif pred_str == 'http://ncat.edu/custom/speed':
                new_p.speed = float(obj_str)
              elif pred_str == 'http://ncat.edu/custom/tasks':
                new_p.tasks.append(obj_str)
              elif pred_str == 'http://xmlns.com/foaf/0.1/name':
                if "webid.ttl" in subj:
                  new_p.friends.append(obj_str)
              #increment
              count = count + 1

      # print ('-------------- the person ---------------')
      # print (new_p.name, ': ', end='' )
      # print (new_p)

      network.append(new_p)

  # create a node with that name and those properties
  # add all nodes
  # then add edges in a separate loop - becasuse if we add edges first,
  #  it will create the same node twice, so we add edges after.
  for p in network:
     # print ('speed attribute', p.speed)
     new_graph.add_node(int(p.name), speed=999 , accuracy=p.accuracy, reliability=p.reliability, tasks=p.tasks )

  # add adding all the edges
  for p in network:
    for f in p.friends:
      # add each edge between p and its f
      if int(p.name) != int(f):   # make sure we don't add ourself to our friends
        new_graph.add_edge(int(p.name), int(f))


  return new_graph
  print (new_graph)
````

````python
####### Dr. Esterline's Code VERSION 2 #######################################
# -*- coding: utf-8 -*-
"""
Created on Fri Dec 23 21:01:59 2022

@author: Dr. Albert Esterline
"""
import networkx as nx
from numpy import random

task_list = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']

def decorate_nodes(G):
    """Add randomly generated values for all 4 attribute for all nodes of G.

    For each node n, randomly generate values for attributes accuracy, speed,
    and reliability, for each, from a uniform [0.0, 1.0) distribution. Also,
    for attribute tasks, randomly select a random number in [2,5] of tasks
    from the global task list.

    Parameters
    ----------
    G : NetworkX Graph
        Each node of G has random values assigned to its 4 attributes accuracy,
        speed, reliability, and tasks

    Returns
    ------
    None
    """
    for n in G.nodes():
        # Get 3 random numbers from the uniform distribution [0, 1).
        # Normalize them so their sum is 1.0. Use them for, respectively,
        # the values of the attribues accuracy, speed,  reliability of node n

        accuracy, speed, reliability = random.uniform(size=3)

        G.nodes()[n]['accuracy'] = accuracy
        G.nodes()[n]['speed'] = speed
        G.nodes()[n]['reliability'] = reliability

        # Get a random number between 2 and 5 for the size of the list of tasks
        # assigned to the tasks attribute of n. There are x tasks from
        # task_list randomly chosen without replacement.
        x = random.randint(2,6)
        G.nodes()[n]['tasks'] = \
            list(random.choice(task_list, size=x, replace=False))

#nx.write_gpickle(G, 'G_pickled')

# Generate a network based on P -- where p is ??
####################################################################3
def gen_network(p):
    G = nx.erdos_renyi_graph(30, p) # 30 nodes and p
    decorate_nodes(G)
    return G

#################################################################

def total_rank(d):
    """ Given a dictionary d, return a list of keys of d ordered in
    non-increasing order of their associated values.
    """
    from operator import itemgetter

    ss = sorted(d.items(), key=itemgetter(1), reverse=True)
    return [x for x, _ in ss]

#################################################################

import functools

def tasks_covered(EG, tasks):
    """ Return True iff all the tasks in argument tasks are covered by the task
    lists that are the values of the tasks attributes of the nodes of G."""
    # Get a list lis of the tasks lists of all nodes of G. Return True iff the
    # tasks in argument tasks is a subset of the union of the task lists in lis.
    lis = [EG.nodes()[n]['tasks'] for n in EG.nodes()]
    return set(tasks) <= set(functools.reduce(lambda a, b: set(a) | set(b), lis))

##########################################################

def mean_centrality(G):
    bc = nx.betweenness_centrality(G)

    # Get the normalized eigenvector centrality
    ec = nx.eigenvector_centrality(G)
    sum_ec = sum(ec.values())
    ec_normalized = {k : v/sum_ec for k,v in ec.items()}

    # Return the harmonic mean of the betweenness centrality and eigenvector
    # centrality for each node
    return {k: (bc[k]*ec_normalized[k])**0.5 for k in ec_normalized.keys()}


def relative_rank(n, sorted_nodes):
    """ Return the rank of n in sorted_nodes. The highest rank is 1.0, the
    lowest, 0.0.
    """
    return 1.0 - sorted_nodes.index(n) / (len(sorted_nodes) - 1)

def average_rank(EG, sorted_nodes):
    return sum([relative_rank(n, sorted_nodes) for n in EG.nodes()]) / len(EG)

def combed_centrality_rank(EG, G, n):
    sorted_nodes = total_rank(mean_centrality(G))
    return (relative_rank(n,sorted_nodes) + average_rank(EG,sorted_nodes)) / 2

#########################################################

# The following function returns the core number of node n in Graph G
# normalized by dividing by the largest core number of any node in G.
def relative_core_num(G, n, max_core_num):
    """ Return the core number of node n in Graph G normalized by dividing by
    the largest core number of any node in G."""
    CN = nx.core_number(G)
    return CN[n] / max_core_num

# The following functions return, respectively, the average core number of the
# nodes in Graph EG and the minimum core number of any node in EG, both
# normalized by dividing by the largest core number of any node in G.
def average_k(EG, G, max_core_num):
    """ Return the average core number of the nodes in Graph EG normalized by
    dividing by the largest core number of any node in G. """
    CN = nx.core_number(EG)
    return (sum(CN.values()) / len(CN)) / max_core_num

def min_k(EG, G, max_core_num):
    """ Return the minimum core number of the nodes in Graph EG normalized by
    dividing by the largest core number of any node in G. """
    return min(nx.core_number(EG).values()) / max_core_num

def combed_core_num(EG, G, n):
    max_core_num = max(nx.core_number(G).values())
    if max_core_num == 0: max_core_num = 1
    rel = relative_core_num(G, 0, max_core_num)
    ave = average_k(EG, G, max_core_num)
    mn  = min_k(EG, G, max_core_num)
    return (rel + ave + mn) / 3.0

#########################################################
import math

def min_max_all_attrs(G, attrs):
    def min_max_attr(G, attr):
        mn, mx = 1, 0
        present = False
        for n in G.nodes():
            if attr in G.nodes()[n]:
                present = True
                val = G.nodes()[n][attr]
                if val > mx: mx = val
                if val < mn: mn = val
        if not present:
            raise Exception('Attribute ' + attr +
                            ' not present in the network')
        return mn, mx

    return {attr: min_max_attr(G, attr) for attr in attrs}

def ave_all_attr_values_norm(EG, attrs, d):
    def ave_attr_value_norm(EG, attr, mx):
        sm, cnt = 0.0, 0
        for n in EG.nodes():
            if attr in EG.nodes()[n]:
                sm += EG.nodes()[n][attr]
                cnt += 1
        return (sm/cnt)/mx if cnt > 0 else 0

    return sum([ave_attr_value_norm(EG, attr, d[attr][1]) for attr in attrs])\
           / len(attrs)

def g_mean_min_attr_values_norm(EG, attrs, d):
    def min_attr_value_norm(EG, attr, mn, mx):
        mn1 = 1.0
        for n in EG.nodes():
            if attr in EG.nodes()[n]:
                val = EG.nodes()[n][attr]
                if val < mn1: mn1 = val
        # Guard against multiplication by 0 with a very small value instead of 0
        #diff = 0.0005 if mn1 == mn else mn1 - mn
        #print((0.003 + 0.997 * (diff / (mx - mn))) ** (1/5))
        return 1.0 if mx == mn else (0.003 + 0.997 * ((mn1 - mn) / (mx - mn))) ** (1/5)

    return math.prod([min_attr_value_norm(EG, attr, *d[attr])
                      for attr in attrs]) ** (1/len(attrs))
    # return len(attrs) / \
    #     sum([1 / min_attr_value_norm(EG, attr, *d[attr]) for attr in attrs])

def ave_min_all_attr_values_norm(EG, G, attrs):
    global_dict = min_max_all_attrs(G, attrs)
    return ave_all_attr_values_norm(EG, attrs, global_dict), \
           g_mean_min_attr_values_norm(EG, attrs, global_dict)


#########################################################

def average_assortativity(EG, attrs):
    """ Return the average of the assortativity for EG over attributes accuracy,
    speed, and reliability"""
    # Get a list of the assortativity of G for attributes accuracy, speed, and
    # reliability, and return their average.
    assort = [nx.numeric_assortativity_coefficient(EG,x) for x in attrs]
    return ((sum(assort) / 3.0) + 1) / 2

##############################################################

# The following functions return, respectively, the average degree of the
# nodes in graph G and the minimum degree of any node in G.
def average_degree(EG, G, max_deg):
    """ Return the normalized average degree of the nodes in Graph EG. """
    return (sum(dict(G.degree(nbunch=EG.nodes())).values()) / len(EG)) / max_deg

def min_degree(EG, max_deg):
    """ Return the normalized minimum cdegree of the nodes in Graph EG. """
    return min(dict(EG.degree()).values()) / max_deg

def combed_degree(EG, G):
    max_deg = max(dict(G.degree()).values())
    if max_deg == 0: max_deg = 1
    return (average_degree(EG, G, max_deg) + min_degree(EG, max_deg)) / 2.0

###############################################################

def normed_combed_deg_assort(G, EG):
    deg_assort_full = nx.degree_assortativity_coefficient(G)
    deg_assort_embed_ego = \
        nx.degree_assortativity_coefficient(G, nodes=EG.nodes())
    deg_assort_ego = nx.degree_assortativity_coefficient(EG)
    deg_assort_embed_ego_norm = (deg_assort_embed_ego - deg_assort_full) /2
    deg_assort_ego_norm = (deg_assort_ego - deg_assort_full) /2
    return (((deg_assort_embed_ego_norm + deg_assort_ego_norm) / 2.0) + 1) / 2

###################################################################

def overall(EG, G, n):
    attrs = ['accuracy', 'speed', 'reliability']
    ave_perf, min_perf = ave_min_all_attr_values_norm(EG, G, attrs)
    return 0.35 * combed_centrality_rank(EG, G, n) \
         + 0.23 * combed_core_num(EG, G, n)        \
         + 0.15 * ave_perf \
         + 0.09 * min_perf \
         + 0.07 * average_assortativity(EG, attrs)        \
         + 0.06 * combed_degree(EG, G)             \
         + 0.04 * normed_combed_deg_assort(G, EG)

##################################################################
import statistics
import numpy as np

#bkd --- store the networks we make here, in this list
glist = []

def global_ranking(G, tasks):
    fails = 0
    ranking = {}
    for n in G.nodes():
        EG = nx.ego_graph(G, n)
        if tasks_covered(EG, tasks):
            ranking[n] = round(overall(EG, G, n), 3)
        else:
            print("Node", n, " and its neighbors don't handle all tasks.")
            fails += 1
    return ranking, fails


# Run Trials - this may generate the networks
# - start
# - stp
# - numb ??
# - repititions?

def run_trials(strt, stp, num, reps):
    ps, ave_mins, ave_maxs, ave_aves, ave_fails = [], [], [], [], []
    folder_num = 0
    for p in np.linspace(strt, stp, num=num):
        mins, maxs, aves, fails = [], [], [], []
        # for count in Repititions - for each reptition
        for cnt in range(reps):
            # the original code generated a network here
            # we will read it from file
            # G = gen_network(p)
            # manually calc folder name
            folder_name = str(folder_num).zfill(3)
            G = read_in_a_network(folder_name)

            rnking, fls = global_ranking(G, {'B', 'D', 'F'})
            fails.append(fls)
            vals = sorted(rnking.values())
            mins.append(vals[0])
            maxs.append(vals[-1])
            aves.append(statistics.mean(vals))
        ps.append(p)
        ave_fails.append(statistics.mean(fails))
        ave_mins.append(statistics.mean(mins))
        ave_maxs.append(statistics.mean(maxs))
        ave_aves.append(statistics.mean(aves))
    return ps, ave_mins, ave_maxs, ave_aves, ave_fails

pp, a_mins, a_maxs, a_aves, a_fails = run_trials(0.2, 1.0, 25, 12)

import matplotlib.pyplot as plt
# plt.plot(pp, a_mins, 'k^',  label='Minimum')
# plt.plot(pp, a_aves, 'k-',  label='Average')
# plt.plot(pp, a_maxs, 'k--', label='Maximum')
# plt.legend(fontsize=8)
# plt.xlabel('p (probability of an edge between any 2 nodes)')
# plt.ylabel('score', fontsize=15)
# plt.title('Score vs. p', fontsize=16)
# plt.figure()
plt.plot(pp, a_fails)
plt.xlabel('p (probability of an edge between any 2 nodes)')
plt.ylabel('Average failures for 45 nodes')
plt.title('Failures to cover all tasks vs. p')

# for i = 1 to 25
# for x in range(25):
#   run a round pp[x], 3
#    print(round(pp[x],3), round(a_mins[x],3), round(a_aves[x],3), round(a_maxs[x],3))

# delta_p =  0.7 - 0.332
# delta_min = 0.451  - 0.385
# delta_ave = 0.563 - 0.511
# delta_max = 0.676 - 0.629
# print(round(delta_min/delta_p, 3), round(delta_ave/delta_p, 3),
#       round(delta_max/delta_p, 3))

# def intersct(m, p0, y0):
#     return round(y0 - m * p0, 3)
# intersct(0.179, 0.332, 0.385)
# intersct(0.141, 0.332, 0.511)
# intersct(0.128, 0.332, 0.629)

# f = open('temp_vals_pp_v3.txt', 'w')
# f.write(str(pp))
# f.close()

# f = open('temp_vals_mins_v3.txt', 'w')
# f.write(str(a_mins))
# f.close()

# f = open('temp_vals_aves_v3.txt', 'w')
# f.write(str(a_aves))
# f.close()

# f = open('temp_vals_maxs_v3.txt', 'w')
# f.write(str(a_maxs))
# f.close()

# pp1 = pp[3:24]
# a_mins1 = a_mins[3:24]
# a_aves1 = a_aves[3:24]
# a_maxs1 = a_maxs[3:24]
# import pandas as pd
# from statsmodels.formula.api import ols
# data = pd.DataFrame({'x': pp1, 'y': a_mins1})
# model = ols("y ~ x", data).fit()
# print(model.params)
# data1 = pd.DataFrame({'x': pp1, 'y': a_aves1})
# model1 = ols("y ~ x", data1).fit()
# print(model1.params)
# data2 = pd.DataFrame({'x': pp1, 'y': a_maxs1})
# model2 = ols("y ~ x", data2).fit()
# print(model2.params)
````
